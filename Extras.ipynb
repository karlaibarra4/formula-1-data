## 2023 schedule

```python
e1 = pd.read_html('https://en.wikipedia.org/wiki/2023_Formula_One_World_Championship')
print(e1[1])
```

```python
de1 = pd.DataFrame(e1[1])
de1
```

```python
de1.columns = ['round', 'grandPrix', 'circuitName', 'date']
de1 = de1.drop(23)
de1
```

```python
de1.loc[:,'season'] = ['2023']*23
de1 = de1.reindex(columns = ['season', 'round', 'grandPrix', 'circuitName', 'date'])
de1
```

```python
pd.DataFrame(de1).to_csv('schedule2023.csv', index = False)
```

## Circuits history

```python
url = 'https://en.wikipedia.org/wiki/List_of_Formula_One_circuits'
content = pd.read_html(url)
df = pd.DataFrame(content[2])
```

```python
df
```

```python
df.drop(['Map', 'Type', 'Direction', 'Last length used'], axis = 'columns', inplace=True)
```

```python
df
```

```python
df['active'] = df['Season(s)'].apply(lambda x: x[:4])
df
```

```python
df.columns = ['circuitName', 'location', 'country', 'raceId', 'period', 'prixHeld', 'active']
df[3::]
```

```python
pd.DataFrame(df).to_csv('CircuitActive.csv', index = False)
```
