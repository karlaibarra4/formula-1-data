```python
import requests
import json
import pandas as pd
import warnings
warnings.filterwarnings("ignore")
```

## Tabla de Lap Times

```python
url = "http://ergast.com/api/f1/2011/5/laps/1.json"
response = requests.get(url)
content = json.loads(response.content)

content['MRData']['RaceTable']['Races']
```

```python
lap = {'season':[],
       'round':[],
       'raceId':[],
       'driverId':[],
       'number':[],
       'position':[],
       'time':[]
      }
```

```python
#Número de carreras que hubo en cada temporada +1 ya que toma el límite superior como un si fuera un conjunto abierto
carreras={int('2010'):20,
          int('2011'):20,
          int('2012'):21,
          int('2013'):20,
          int('2014'):20,
          int('2015'):20,
          int('2016'):22,
          int('2017'):21,
          int('2018'):22,
          int('2019'):22,
          int('2020'):18,
          int('2021'):23,
          int('2022'):23
         }
list(range(1,carreras[2022]))
```

```python
for year in list(range(2010,2023)):
    urlap = 'http://ergast.com/api/f1/{}'
    urlap1 = urlap.format(year)
    for race in list(range(1,carreras[year])):
        urlrace = '/{}/laps.json'
        urlrace1 = urlrace.format(race)
        urlt = urlap1 + urlrace1
        responselaps = requests.get(urlt)
        jsonlaps = responselaps.json()    
        for i in jsonlaps['MRData']['RaceTable']['Races']:
            for time1 in i['Laps']:
                for timings in time1['Timings']:
                    lap['season'].append(int(i['season']))
                    lap['round'].append(int(i['round']))
                    lap['raceId'].append(i['raceName'])
                    lap['driverId'].append(timings['driverId'])
                    lap['number'].append(time1['number'])
                    lap['position'].append(timings['position'])
                    lap['time'].append(timings['time'])
```

```python
pd.DataFrame(lap)
```

```python
pd.DataFrame(lap).to_csv('LapTimes.csv', index = False)
```

## Tabla Driver Standings

```python
url = "http://ergast.com/api/f1/2008/5/driverStandings.json"
response = requests.get(url.format(2008))
content = json.loads(response.content)

content['MRData']['StandingsTable']['StandingsLists'][0]['DriverStandings']
```

```python
dstandings = {'driverId':[],
             'points':[],
             'position':[],
             'positionText':[],
             'wins':[]
             }
```

```python
for year in list(range(2010,2023)):
    url = 'http://ergast.com/api/f1/{}/driverStandings.json'
    response = requests.get(url.format(year))
    content = json.loads(response.content)

    for item in content['MRData']['StandingsTable']['StandingsLists'][0]['DriverStandings']:
        dstandings['driverId'].append(item['Driver']['driverId'])
        dstandings['points'].append(item['points'])
        dstandings['position'].append(int(item['position']))
        dstandings['positionText'].append(item['positionText'])
        dstandings['wins'].append(int(item['wins']))
```

```python
pd.DataFrame(dstandings)
```

```python
pd.DataFrame(dstandings).to_csv('DriverStandings.csv', index = False)
```

## Tabla de Constructor Standings

```python
url = "http://ergast.com/api/f1/2008/5/constructorStandings.json"
response = requests.get(url.format(2008))
content = json.loads(response.content)

content['MRData']['StandingsTable']
```

```python
cstandings = {'constructorId':[],
             'points':[],
             'position':[],
             'positionText':[],
             'wins':[]
             }
```

```python
for year in list(range(2010,2023)):
    url = 'http://ergast.com/api/f1/{}/constructorStandings.json'
    response = requests.get(url.format(year))
    content = json.loads(response.content)

    for item in content['MRData']['StandingsTable']['StandingsLists'][0]['ConstructorStandings']:
        cstandings['constructorId'].append(item['Constructor']['constructorId'])
        cstandings['points'].append(float(item['points']))
        cstandings['position'].append(int(item['position']))
        cstandings['positionText'].append(item['positionText'])
        cstandings['wins'].append(int(item['wins']))
```

```python
pd.DataFrame(cstandings)
```

```python
pd.DataFrame(cstandings).to_csv('ConstructorStandings.csv', index = False)
```
