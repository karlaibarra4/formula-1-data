## Librerias

```python
import requests
import json
import pandas as pd
import warnings
warnings.filterwarnings("ignore")
```

## Tabla de Drivers

```python
url = "http://ergast.com/api/f1/2010/drivers.json"
response = requests.get(url.format(2010))
content = json.loads(response.content)

content['MRData']['DriverTable']
```

```python
drivers = {'season':[],
          'driverId':[],
           'number':[],
           'code':[],
          'forename':[],
          'surname':[],
          'dateOfBirth':[],
          'nationality':[],
          'url':[]}
```

```python
for year in list(range(2010,2023)):
    url = "http://ergast.com/api/f1/{}/drivers.json"
    response = requests.get(url.format(year))
    content = json.loads(response.content)
    path = content['MRData']['DriverTable']
    
    for item in range(0,len(path['Drivers'])):
        drivers['season'].append(int(path['season']))
        drivers['driverId'].append(path['Drivers'][item]['driverId'])
        try:
            drivers['number'].append(path['Drivers'][item]['permanentNumber'])
        except:
            drivers['number'].append(None)
        drivers['code'].append(path['Drivers'][item]['code'])
        drivers['forename'].append(path['Drivers'][item]['givenName'])
        drivers['surname'].append(path['Drivers'][item]['familyName'])
        drivers['dateOfBirth'].append(path['Drivers'][item]['dateOfBirth'])
        drivers['nationality'].append(path['Drivers'][item]['nationality'])
        drivers['url'].append(path['Drivers'][item]['url'])
```

```python
pd.DataFrame(drivers)
```

```python
pd.DataFrame(drivers).to_csv('drivers.csv', index = False)
```

## Tabla de Qualifying Results

```python
url = 'http://ergast.com/api/f1/2022/qualifying.json'
response = requests.get(url.format(2022))
content = json.loads(response.content)
    
content['MRData']['RaceTable']['Races']
```

```python
qualifying = {'season':[],
              'round':[],
              'raceId':[],
              'driverId':[],
              'constructorId':[],
              'number':[],
              'position':[],
              'Q1':[],
              'Q2':[],
              'Q3':[]
}
```

```python
for year in list(range(2010,2023)):
    url = 'https://ergast.com/api/f1/{}/qualifying.json'
    response = requests.get(url.format(year))
    content = json.loads(response.content)
        
    for item in content['MRData']['RaceTable']['Races'][0]['QualifyingResults']:
        qualifying['season'].append(int(content['MRData']['RaceTable']['Races'][0]['season']))
        qualifying['round'].append(int(content['MRData']['RaceTable']['Races'][0]['round']))
        qualifying['raceId'].append(content['MRData']['RaceTable']['Races'][0]['raceName'])
        qualifying['driverId'].append(item['Driver']['driverId'])
        qualifying['constructorId'].append(item['Constructor']['constructorId'])
        qualifying['number'].append(int(item['number']))
        qualifying['position'].append(int(item['position']))
        qualifying['Q1'].append(item['Q1'])
        try:
            qualifying['Q2'].append(item['Q2'])
        except:
            qualifying['Q2'].append(None)
        try: 
            qualifying['Q3'].append(item['Q3'])
        except:
            qualifying['Q3'].append(None)
```

```python
pd.DataFrame(qualifying)
```

```python
pd.DataFrame(qualifying).to_csv('qualifying.csv', index = False)
```
