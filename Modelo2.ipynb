## Tabla de Sprint Qualifying Results

```python
url = "http://ergast.com/api/f1/2021/10/sprint.json"
response = requests.get(url.format(2021))
content = json.loads(response.content)

content['MRData']['RaceTable']['Races']
```

```python
sprint = {'raceId':[],
          'driverId':[],
          'constructorId':[],
          'number':[],
          'position':[],
          'positionText':[],
          'grid':[],
          'points':[],
          'laps':[],
          'fastestLap':[],
          'fastestLapTime':[],
          'status':[],
         }
```

```python
url = 'https://ergast.com/api/f1/2021/10/sprint.json'
response = requests.get(url.format(2021))
content = json.loads(response.content)
        
for item in content['MRData']['RaceTable']['Races']:
    for r in item['SprintResults']:
        sprint['raceId'].append(item['raceName'])
        sprint['driverId'].append(r['Driver']['driverId'])
        sprint['constructorId'].append(r['Constructor']['constructorId'])
        sprint['number'].append(int(r['number']))
        sprint['position'].append(int(r['position']))
        sprint['positionText'].append(r['positionText'])
        sprint['grid'].append(int(r['grid']))
        sprint['points'].append(r['points'])
        sprint['laps'].append(int(r['laps']))
        sprint['fastestLap'].append(int(r['FastestLap']['lap']))
        sprint['fastestLapTime'].append(r['FastestLap']['Time']['time'])
        sprint['status'].append(r['status'])
```

```python
pd.DataFrame(sprint)
```

```python
pd.DataFrame(sprint).to_csv('sprint.csv', index = False)
```

## Tabla de Status

```python
url = "http://ergast.com/api/f1/2022/status.json"
response = requests.get(url.format(2022))
content = json.loads(response.content)

content['MRData']['StatusTable']
```

```python
status = {'season':[],
          'statusId':[],
          'count':[],
          'status':[]
         }
```

```python
for year in list(range(2010,2023)):
    url = 'http://ergast.com/api/f1/{}/status.json'
    response = requests.get(url.format(year))
    content = json.loads(response.content)
        
    for item in content['MRData']['StatusTable']['Status']:
        status['season'].append(int(year))
        status['statusId'].append(int(item['statusId']))
        status['count'].append(int(item['count']))
        status['status'].append(item['status'])
```

```python
pd.DataFrame(status)
```

## Tabla de Results

```python
url = "http://ergast.com/api/f1/2008/5/results.json"
response = requests.get(url.format(2008))
content = response.json()

content['MRData']['RaceTable']['Races']
```

```python
results = {'season':[],
           'raceId':[],
           'driverId':[],
           'constructorId':[],
           'number':[],
           'grid':[],
           'position':[],
           'positionText':[],
           'points':[],
           'laps':[],
           'statusId':[],
           'time':[],
           'milliseconds':[],
           'fastestLap':[],
           'fastestLapRank':[],
           'fastestLapTime':[],
           'fastestLapSpeed':[]
          }
```

```python
for year in list(range(2010,2023)):
    url = 'http://ergast.com/api/f1/{}/results.json'
    response = requests.get(url.format(year))
    content = json.loads(response.content)
        
    for item in content['MRData']['RaceTable']['Races']:
        for r in item['Results']:
            results['season'].append(item['season'])
            results['raceId'].append(item['raceName'])
            results['driverId'].append(r['Driver']['driverId'])
            results['constructorId'].append(r['Constructor']['constructorId'])
            results['number'].append(int(r['number']))
            results['grid'].append(int(r['grid']))
            results['position'].append(int(r['position']))
            results['positionText'].append(r['positionText'])
            results['points'].append(r['points'])
            results['laps'].append(int(r['laps']))
            results['statusId'].append(r['status'])
            try:
                results['milliseconds'].append(r['Time']['millis'])
            except:
                results['milliseconds'].append(None)

            try:
                results['time'].append(r['Time']['time'])
            except:
                results['time'].append(None)

            try:
                results['fastestLapRank'].append(r['FastestLap']['rank'])
            except:
                results['fastestLapRank'].append(None)

            try:
                results['fastestLap'].append(r['FastestLap']['lap'])
            except:
                results['fastestLap'].append(None)

            try:
                results['fastestLapTime'].append(r['FastestLap']['Time']['time'])
            except:
                results['fastestLapTime'].append(None)

            try:
                results['fastestLapSpeed'].append(r['FastestLap']['AverageSpeed']['speed'])
            except:
                results['fastestLapSpeed'].append(None)
```

```python
pd.DataFrame(results)
```

```python
pd.DataFrame(results).to_csv('results.csv', index = False)
```

## Tabla de Pit Stops

```python
url = "http://ergast.com/api/f1/2011/5/pitstops.json"
response = requests.get(url)
content = response.json()

content['MRData']['RaceTable']['Races']
```

```python
pitstops = {'season':[],#season
            'round':[],#round
            'raceId':[],#raceName
            'circuitId':[],
            'driverId':[],#driverId
            'lap':[],#lap
            'stop':[],#stop
            'time':[],
            'duration':[]
           }
```

```python
#Número de carreras que hubo en cada temporada +1 ya que toma el límite superior como un si fuera un conjunto abierto
carreras = {int('2010'):20,
          int('2011'):20,
          int('2012'):21,
          int('2013'):20,
          int('2014'):20,
          int('2015'):20,
          int('2016'):22,
          int('2017'):21,
          int('2018'):22,
          int('2019'):22,
          int('2020'):18,
          int('2021'):23,
          int('2022'):23
         }
list(range(1,carreras[2022]))
```

```python
for year in list(range(2010,2023)):
    url='http://ergast.com/api/f1/{}'
    url1 = url.format(year)
    for season in list(range(1,carreras[year])):
        url2='/{}/pitstops.json'
        url3= url2.format(season)
        url4=url1+ url3        
        responseps = requests.get(url4)
        jsonpitstops=responseps.json()
        for i in jsonpitstops['MRData']['RaceTable']['Races']:
            for ps in i['PitStops']:
                try:
                    pitstops['season'].append(int(i['season']))
                    pitstops['round'].append(int(i['round']))
                    pitstops['raceId'].append(i['raceName'])
                    pitstops['circuitId'].append(i['Circuit']['circuitId'])
                    pitstops['driverId'].append(ps['driverId'])
                    pitstops['lap'].append(ps['lap'])
                    pitstops['stop'].append(ps['stop'])
                    pitstops['time'].append(ps['time'])
                    pitstops['duration'].append(ps['duration'])
                except:
                    pitstops['driverId'].append(None)
                    pitstops['lap'].append(None)
                    pitstops['stop'].append(None)
                    pitstops['time'].append(None)
                    pitstops['duration'].append(None)
```

```python
pd.DataFrame(pitstops)
```

```python
pd.DataFrame(pitstops).to_csv('pitstops.csv', index = False)
```
