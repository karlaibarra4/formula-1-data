```python
races = {'season':[],
        'round':[],
        'raceName':[],
        'circuitId':[],
        'long':[],
        'country':[],
        'url':[]}
```

```python
for year in list(range(2010,2022)):
    url = 'https://ergast.com/api/f1/{}.json'
    response = requests.get(url.format(year))
    content = json.loads(response.content)
    
    for item in content['MRData']['RaceTable']['Races']:
        races['season'].append(int(item['season']))
        races['round'].append(int(item['round']))
        races['raceName'].append(int(item['raceName']))
        races['circuitId'].append(int(item['circuitId']))
        races['long'].append(int(item['long']))
        races['country'].append(int(item['country']))
        races['url'].append(int(item['url']))
```
